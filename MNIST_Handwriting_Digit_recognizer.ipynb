```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras


(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data() ## Keras has an inbuilt datasets which we can use
```

```python
x_train.shape
```

60000 images, each having dimension (28*28) , it means that
every image has 784 features.

784 features = 1 image, and we have 60000 of such image

```python
y_train=y_train.reshape(60000,1)
```

```python
y_train
```

Displaying an image

```python
plt.imshow(x_train[1])
```

We should flatten the features into a single line as Neural Networks, work on features in linear manner .

```python
x_train=np.reshape(x_train,(60000,28*28))

x_train.shape
```

```python
import tensorflow as tf

# Assuming x_train and y_train are your input data and labels

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(10, activation='softmax'))  # Assuming 10 classes for digit classification
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1000)
```

```python
x_test=x_test.reshape(10000,28*28)
x_test.shape
```

```python
y_pred=model.predict (x_test) ## Predicting the values
```

```python
y_pred=np.argmax(y_pred,axis=1)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Assuming y_test and y_pred are already defined
conf_matrix = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')

# Setting labels and title
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
plt.title('Confusion Matrix')

# Show plot
plt.show()
```
